```
# Some basic setup:
# Setup detectron2 logger
import detectron2
from detectron2.utils.logger import setup_logger
setup_logger()

# import some common libraries
from glob import glob
import numpy as np
import os, json, cv2, random
from google.colab.patches import cv2_imshow

# import some common detectron2 utilities
from detectron2 import model_zoo
from detectron2.engine import DefaultPredictor
from detectron2.config import get_cfg
from detectron2.utils.visualizer import Visualizer
from detectron2.data import MetadataCatalog, DatasetCatalog

from PIL import Image
import argparse
from pathlib import Path
```

```
config_file = 'COCO-Detection/faster_rcnn_R_50_FPN_3x.yaml'
```

```
cfg = get_cfg()

cfg.merge_from_file(model_zoo.get_config_file(config_file))
cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.5  # set threshold for this model

cfg.MODEL.WEIGHTS = model_zoo.get_checkpoint_url(config_file)
predictor = DefaultPredictor(cfg)
```

```
images_path = './taskC_files/images_to_test/'
object_path = './taskC_files/objects_to_insert/'
```

```
for image in os.listdir(images_path):
  print(image)
```

```
case = 'sheep'
```

```
for image in os.listdir(images_path + case):
  print(image)
  im = np.array(Image.open(images_path + case + '/' + image))
  outputs = predictor(im)

  v = Visualizer(im[:, :, ::-1], MetadataCatalog.get(cfg.DATASETS.TRAIN[0]), scale=1)
  out = v.draw_instance_predictions(outputs["instances"].to("cpu"))
  image = Image.fromarray(out.get_image()[:, :, ::-1])
  image.save(f'./results/co-occurring/{case}.png') 
```
